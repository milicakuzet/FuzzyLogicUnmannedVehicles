```python
#Cilj je u 2D-u predstaviti putanju bespilotne letelice od pocetnog stanja O, do krajnje pozicije T
#ulaz je brzina letelice i ugao kretanja
#tau_v i tau_theta nam predstavljaju vremenske parametre
```

```python
import numpy as np
```

```python
from scipy.integrate import odeint
```

```python
from matplotlib import pyplot as plt
```

```python
tau_v=5.0 #vremensko kasnjenje za brzinu
tau_theta=2.0 #vremensko kasnjenje za ugao
v_c= 10.0 #brzina
theta_c = np.pi / 4 #ugao 
```

```python
#poziciju same letelice odredjujemo preko formula
# x = v * cos(theta)
# y = v * sin(theta)
```

```python
def model(state, t):
    x,y,v,theta = state
    v_dot = (v_c - v) / tau_v
    theta_dot = (theta_c - theta) / tau_theta
    x_dot = v * np.cos(theta)
    y_dot = v * np.sin(theta)
    return [x_dot, y_dot, v_dot, theta_dot]
```

```python
initial_state = [0.0, 0.0, 0.0, 0.0]
```

```python
t= np.linspace(0,50,500)
```

```python
#funkcija za resavanje sistema diferencijalnih jednacina
solution = odeint(model, initial_state,  t)
```

```python
x= solution[:, 0]
y= solution[:, 1]
v= solution[:, 2]
theta= solution[:, 3]
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y,label='Putanja UAV-a')
plt.xlabel('x pozicija')
plt.ylabel('y pozicija')
plt.title('Simulacija putanje UAV-a')
plt.legend()
plt.grid(True)
plt.show()
```
