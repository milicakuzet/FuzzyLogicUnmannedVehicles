```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Definisanje mreže
grid_size = 100  # dimenzije mreže
obstacle_density = 0.1  # gustina prepreka (10%)
num_obstacles = int(obstacle_density * grid_size**2)

# Postavljanje prepreka nasumično u mrežu
np.random.seed(0)  # zbog reprodukovanja rezultata
obstacles_x = np.random.randint(0, grid_size, num_obstacles)
obstacles_y = np.random.randint(0, grid_size, num_obstacles)
obstacles = list(zip(obstacles_x, obstacles_y))

# Definisanje fuzzy ulaza
distance_to_obstacle = ctrl.Antecedent(np.arange(0, 101, 1), 'distance_to_obstacle')
angle_to_obstacle = ctrl.Antecedent(np.arange(-180, 181, 1), 'angle_to_obstacle')

# Definisanje fuzzy izlaza
speed_change = ctrl.Consequent(np.arange(-100, 101, 1), 'speed_change')
heading_change = ctrl.Consequent(np.arange(-45, 46, 1), 'heading_change')

# Funkcije pripadnosti za udaljenost
distance_to_obstacle['close'] = fuzz.trimf(distance_to_obstacle.universe, [0, 0, 30])
distance_to_obstacle['medium'] = fuzz.trimf(distance_to_obstacle.universe, [20, 50, 80])
distance_to_obstacle['far'] = fuzz.trimf(distance_to_obstacle.universe, [50, 100, 100])

# Funkcije pripadnosti za ugao
angle_to_obstacle['neg_big'] = fuzz.trimf(angle_to_obstacle.universe, [-180, -180, -120])
angle_to_obstacle['neg_med'] = fuzz.trimf(angle_to_obstacle.universe, [-120, -90, -30])
angle_to_obstacle['neg_small'] = fuzz.trimf(angle_to_obstacle.universe, [-30, -10, 0])
angle_to_obstacle['pos_small'] = fuzz.trimf(angle_to_obstacle.universe, [0, 10, 30])
angle_to_obstacle['pos_med'] = fuzz.trimf(angle_to_obstacle.universe, [30, 90, 120])
angle_to_obstacle['pos_big'] = fuzz.trimf(angle_to_obstacle.universe, [120, 180, 180])

# Funkcije pripadnosti za promenu brzine i pravca
speed_change['very_slow'] = fuzz.trimf(speed_change.universe, [-100, -100, -50])
speed_change['slow'] = fuzz.trimf(speed_change.universe, [-50, 0, 50])
speed_change['fast'] = fuzz.trimf(speed_change.universe, [50, 100, 100])

heading_change['neg_big'] = fuzz.trimf(heading_change.universe, [-45, -45, -15])
heading_change['neg_med'] = fuzz.trimf(heading_change.universe, [-15, -10, 0])
heading_change['neg_small'] = fuzz.trimf(heading_change.universe, [-5, 0, 0])
heading_change['zero'] = fuzz.trimf(heading_change.universe, [-1, 0, 1])
heading_change['pos_small'] = fuzz.trimf(heading_change.universe, [0, 5, 15])
heading_change['pos_med'] = fuzz.trimf(heading_change.universe, [10, 15, 45])
heading_change['pos_big'] = fuzz.trimf(heading_change.universe, [15, 45, 45])

# Pravila za fuzzy logiku
rule1 = ctrl.Rule(distance_to_obstacle['far'] & angle_to_obstacle['pos_med'], (speed_change['fast'], heading_change['zero']))
rule2 = ctrl.Rule(distance_to_obstacle['far'] & angle_to_obstacle['neg_med'], (speed_change['fast'], heading_change['zero']))
rule3 = ctrl.Rule(distance_to_obstacle['medium'] & angle_to_obstacle['pos_small'], (speed_change['slow'], heading_change['neg_small']))
rule4 = ctrl.Rule(distance_to_obstacle['close'] & angle_to_obstacle['neg_big'], (speed_change['very_slow'], heading_change['pos_big']))

# Kreiranje kontrolnog sistema
uav_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4])
uav_sim = ctrl.ControlSystemSimulation(uav_ctrl)

# Parametri za simulaciju kretanja letelice
x, y = [0], [0]  # početne koordinate letelice
current_speed = 50  # početna brzina letelice
current_heading = 0  # početni ugao kretanja letelice
goal = (grid_size - 1, grid_size - 1)  # cilj je donji desni ugao mreže

# Funkcija za izračunavanje udaljenosti i ugla do prepreke
def calc_distance_angle(x, y, obstacle_x, obstacle_y):
    distance = np.sqrt((x - obstacle_x) ** 2 + (y - obstacle_y) ** 2)
    angle = np.degrees(np.arctan2(obstacle_y - y, obstacle_x - x)) - current_heading
    return distance, angle

# Simulacija kretanja letelice
while (x[-1], y[-1]) != goal:
    # Pronalazak najbliže prepreke
    distances_angles = [calc_distance_angle(x[-1], y[-1], ox, oy) for ox, oy in obstacles]
    closest_obstacle = min(distances_angles, key=lambda t: t[0])

    # Postavljanje ulaza za fuzzy sistem
    uav_sim.input['distance_to_obstacle'] = closest_obstacle[0]
    uav_sim.input['angle_to_obstacle'] = closest_obstacle[1]

    # Izračunavanje fuzzy izlaza
    uav_sim.compute()

    # Provera da li su izlazi validni
    current_speed_change = uav_sim.output.get('speed_change', None)
    current_heading_change = uav_sim.output.get('heading_change', None)

    if current_speed_change is None or current_heading_change is None:
        print(f"Greška: Fuzzy sistem nije uspeo da generiše izlaze za trenutnu iteraciju.")
        print(f"Ulazi: Udaljenost do prepreke: {closest_obstacle[0]}, Ugao: {closest_obstacle[1]}")
        break

    # Ažuriranje brzine i pravca
    current_speed += current_speed_change / 10
    current_heading += current_heading_change

    # Ažuriranje pozicije letelice
    x_new = x[-1] + current_speed * np.cos(np.radians(current_heading))
    y_new = y[-1] + current_speed * np.sin(np.radians(current_heading))

    # Ograničavanje da letelica ostane u mreži
    x_new = min(max(x_new, 0), grid_size - 1)
    y_new = min(max(y_new, 0), grid_size - 1)

    x.append(x_new)
    y.append(y_new)

# Crtanje putanje i prepreka
plt.figure(figsize=(8, 8))

# Iscrtavanje prepreka
for ox, oy in obstacles:
    plt.plot(ox, oy, 'rx', markersize=12)

# Iscrtavanje putanje letelice
plt.plot(x, y, label="Putanja letelice", color="blue", marker="o", markersize=5, linestyle="-")
plt.xlim(0, grid_size)
plt.ylim(0, grid_size)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Putanja letelice kroz mrežu sa preprekama')
plt.grid(True)
plt.show()
```
